```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# Categorical Plots

# Categorical Scatter Plot
    # Stripplot
    # Swarmplot

# Categorical Distribution Plots
    # Boxplot
    # Violinplot

# Categorical Estimate Plot -> for central tendency
    # Barplot
    # Pointplot
    # Countplot

#Figure level function -> catplot
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
# strip plot
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill', jitter=False)
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# Using figure level function
# catplot
sns.catplot(data=tips, x='day',y='total_bill',kind='strip')
```

```python
# jitter
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```

```python
# swarmplot
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm')
```

```python
sns.swarmplot(data=tips, x='day',y='total_bill', hue='sex')
```

## Boxplot

```python
# A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). 
# It can tell you about your outliers and what their values are. 
# Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.
```

```python
# Boxplot
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
# Boxplot
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips,y='total_bill')
```

```python
# Violinplot = (Boxplot + KDEplot)
```

```python
# violinplot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin')
```

```python
# hue

sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
# barplot
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# barplot
# some issue with errorbar
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill',hue='smoker',estimator=np.min)
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', errorbar=None)
```

```python
# point plot
sns.pointplot(data=tips, x='sex', y='total_bill',hue='smoker',errorbar=None)
```

```python
# When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars
```

```python
# countplot
sns.countplot(data=tips, x='sex', hue='day')
```

```python
# A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. 
# This is similar to a histogram over a categorical, rather than quantitative, variable
```

```python
# faceting using catplot
sns.catplot(data=tips, x='sex', y='total_bill', col='smoker', kind='box', row='time')
```

## Regression Plot

```python
# regplot  -> axes level
# lmplot   -> figure level

# I n the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.
```

```python
# axes level
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
# residplot
sns.residplot(data=tips, x='total_bill', y='tip')
```

```python
# A second way to plot Facet plots -> FacetGrid
```

```python
# figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time')
g.map(sns.violinplot, 'sex', 'total_bill', order=['Male', 'Female'])
```

```python
tips.info()
```

```python
g = sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
```

## Plotting Pairwise Relationship (PairGrid Vs PairPlot)

```python
sns.pairplot(data=iris, hue='species')
```

```python
# Pairgrid
g = sns.PairGrid(data=iris, hue='species')

g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

## JointGrid Vs Jointplot

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='hist',hue='sex')
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.histplot)
```
